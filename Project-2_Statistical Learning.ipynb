# 2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

```python
ti = ti.rename(columns={'Old Scheme (in thousands)': 'OldScheme', 'New Scheme (in thousands)': 'NewScheme'})
ti.head()
```

```python
# Null Hypothesis : There is no significant difference between outputs of old scheme and new scheme
# Alternate Hypothesis : There is a significant difference/ New Scheme has significantly raised outputs
```

```python
Old = ti.OldScheme
New = ti.NewScheme
```

```python
from scipy.stats import ttest_1samp,ttest_ind,wilcoxon,levene
t_statistic, p_value = ttest_ind(Old , New)
print("t stat value:",t_statistic)
print("p value :",p_value)                      
```

# 3. What conclusion does the test (p-value) lead to? (2.5 points)

```python
# p_value > 0.05 , thus we cannot reject null hypothesis.
# Accepting Null hypothesis
# There is no significant difference between outputs of old scheme and new scheme
```

```python
# We can conclude that output raised by Old scheme and New scheme are equal.
# OR 
# There is no difference between ouputs raised by Old sheme and New scheme.
```
